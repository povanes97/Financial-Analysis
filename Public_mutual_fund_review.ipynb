```python
# top 50 funds  of 10 yr returns
df.head()
df['10 yr']=pd.to_numeric(df['10 yr'],errors='coerce')
df['10 yr'].dtype
ranking=df.sort_values(by='10 yr',ascending=False)
display(ranking.head(50))
top5010YR=ranking.head(50)

plt.figure(figsize=(12,12))
plt.barh(top5010YR['Fund name'],top5010YR['10 yr'])
plt.xticks(rotation=45)
plt.title("Top 50 Funds VS 10Yr return")
plt.xlabel('10 YR returns')
plt.ylabel('Fund Name')
plt.gca().invert_yaxis()
plt.tight_layout()
plt.savefig('Top 50 VS 10 Yr Return.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
# consistency scrore by definition :Can I stay invested long-term but still have liquidity when needed, without killing returns?
df['rank_YTD']=df['YTD'].rank(ascending=False,na_option='bottom')
df['rank_1Y'] = df['1 yr'].rank(ascending=False,na_option='bottom')
df['rank_3Y'] = df['3 yr'].rank(ascending=False,na_option='bottom')
df['rank_5Y'] = df['5yr'].rank(ascending=False,na_option='bottom')
df['rank_10Y'] = df['10 yr'].rank(ascending=False,na_option='bottom')


df['consistency_score']=(
    df['rank_YTD']+
    df['rank_1Y']+
    df['rank_3Y']+
    df['rank_5Y']+
    df['rank_10Y']
)

top50=df.sort_values('consistency_score').head(50)
display(top50)
top50.to_csv('public_mutual_cleaned.csv',index=False)
import os
print(os.getcwd())
#df.head()
```

```python
#Stability vs growth plotting the scatter plot 

plt.figure(figsize=(10,6))
scatter=plt.scatter(top50['10 yr'],top50['1 yr'],c=top50['consistency_score'],cmap='viridis_r')
cbar=plt.colorbar(scatter,pad=0.15)
cbar.set_label('Consistency Score(Lower=Better)',labelpad=15)

for _,row in top50.iterrows():
    plt.annotate(
        row['Fund name'],
        (row['10 yr'],row['1 yr']),
        fontsize=4)
    
plt.xlabel('10-year Return(Long terrm Investment)')
plt.ylabel('1-year Return(Short term Investment)')
plt.title('Liquidity vs Long-Term Growth Trade-Off(Top 50 Accounts)')
plt.axvline(top50['10 yr'].mean(),linestyle='--')
plt.axhline(top50['1 yr'].mean(),linestyle='--')


plt.tight_layout()
plt.savefig('Liquidity Vs Long-Term Growth.png', dpi=300,bbox_inches='tight')
plt.show()
```
